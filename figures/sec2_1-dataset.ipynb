# Dataset Description

## Describe TPC-DS runtime

```python
import pandas as pd
from pathlib import Path

data_dir = Path("../data/")
assert data_dir.exists()

# Load TPC-DS dataset
small_tpcds = pd.read_csv(data_dir / "TPC-DS-10TB-Snowflake.csv")
large_tpcds = pd.read_csv(data_dir / "TPC-DS-100TB-Snowflake.csv")

# Plot the small tpc-ds TOTAL_ELAPSED_TIME CDF
ax = (small_tpcds["TOTAL_ELAPSED_TIME"] / 1000).plot.hist(
    bins=50, cumulative=True, density=1, 
    histtype='step', linewidth=2,
    title="TPC-DS 10TB Snowflake Execution Time CDF",
    xlabel="Execution time (second)",
    ylabel="CDF",
    legend=True,
)

import matplotlib.pyplot as plt

# Plot the large tpc-ds TOTAL_ELAPSED_TIME CDF
(large_tpcds["TOTAL_ELAPSED_TIME"] / 1000).plot.hist(
    bins=50, cumulative=True, density=1, 
    histtype='step', linewidth=2,
    color="maroon",
    title="TPC-DS 100TB Snowflake Execution Time CDF",
    xlabel="Execution time (second)",
    ylabel="CDF",
    legend=True,
)
plt.savefig("sec2_1-exectime.png")
```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

data_dir = Path("../data/")
assert data_dir.exists()

# Load TPC-DS dataset
small_tpcds = pd.read_csv(data_dir / "TPC-DS-10TB-Snowflake.csv")
large_tpcds = pd.read_csv(data_dir / "TPC-DS-100TB-Snowflake.csv")

# Plot the small tpc-ds TOTAL_ELAPSED_TIME CDF
(small_tpcds["TOTAL_ELAPSED_TIME"] / 1000).plot.hist(
    bins=50, cumulative=True, density=1, 
    histtype='step', linewidth=2,
    title="TPC-DS 10TB Snowflake Execution Time CDF",
    xlabel="Execution time (second)",
    ylabel="CDF",
    legend=True,
)
# Plot the large tpc-ds TOTAL_ELAPSED_TIME CDF
(large_tpcds["TOTAL_ELAPSED_TIME"] / 1000).plot.hist(
    bins=50, cumulative=True, density=1, 
    histtype='step', linewidth=2,
    color="maroon",
    title="TPC-DS 100TB Snowflake Execution Time CDF",
    xlabel="Execution time (second)",
    ylabel="CDF",
    legend=True,
)

# Load TPC-DS dataset
aug_tpcds = pd.read_csv(data_dir / "Augmented-TPC-DS-10TB-Snowflake.csv")

# Plot the small tpc-ds TOTAL_ELAPSED_TIME CDF
(aug_tpcds["TOTAL_ELAPSED_TIME"] / 1000).plot.hist(
    bins=50, cumulative=True, density=1, 
    histtype='step', linewidth=2,
    title="Augmented TPC-DS 10TB Snowflake Execution Time CDF",
    xlabel="Execution time (second)",
    ylabel="CDF",
    legend=True,
)

# Change legend text
plt.legend(["TPC-DS 10TB", "TPC-DS 100TB", "Augmented TPC-DS 10TB"])
plt.savefig("sec3_2-exectime-compare.png")
```
